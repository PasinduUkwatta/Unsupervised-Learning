```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=300
D=2
K=3
```

```python
#cerate the data
mu1=np.array([0,0])
mu2=np.array([5,5])
mu3=np.array([0,5])

X=np.zeros((N,D))
X[:100,:] =np.random.rand(100,D) + mu1
X[100:200,:] =np.random.rand(100,D) + mu2
X[200:,:] =np.random.rand(100,D) + mu3

Y =np.array([0]*100 +[1]*100+[2]*100)
```

```python
#visulaize the data
plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
X.mean(axis=0).shape
```

```python
#find the mean for cluster
means =np.zeros((K,D))

means[0,:] =X[Y==0].mean(axis=0)
means[1,:] =X[Y==1].mean(axis=0)
means[2,:] =X[Y==2].mean(axis=0)
```

```python
#plot the data
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(means[:,0],means[:,1],s=500,c="red",marker="*")
```

```python
from random import randrange

def func(n):
   mydict = {}
   for i in range(n):
     mydict[str(i)] = randrange(3)
   return mydict
 
```
